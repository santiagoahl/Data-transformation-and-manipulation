```python
import pandas as pd
```

```python
path = '/content/bestsellers-with-categories.csv'
df_books = pd.read_csv(path,sep=',',header=0)
df_books.head(2)
```

```python
df_books['Year'] > 2016
```

```python
year_condition = df_books['Year'] > 2016
```

```python
df_books[year_condition]
```

```python
genre_condition = df_books['Genre']=='Fiction'
```

```python
#AND
df_books[genre_condition & year_condition]
```

`Alt Gr + +` = `~
`

```python
#NOT
df_books[~year_condition]
```
