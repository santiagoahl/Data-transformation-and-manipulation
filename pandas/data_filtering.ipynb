So we can change them

```python
cols = ['NAME', 'AUTHOR', 'USER RATING', 'REVIEWS', 'PRICE', 'YEAR', 'GENRE']
```

```python
df_books = pd.read_csv('/content/bestsellers-with-categories.csv', sep=',', header=0, names=cols)
df_books
```

# Filtering

```python
#By names
df_books['NAME']
```

```python
#By more column parameters:
df_books[['NAME','AUTHOR','YEAR']]
```

## [`loc`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.loc.html)

```python
df_books.loc[:]
```

```python
df_books.loc[0:4] #4 is inclusive
```

## Loc slicing + column parameter search

```python
df_books.loc[0:4,['NAME','AUTHOR']]
```

But what if we want to **Manipulate only a row**

```python
df_books.loc[:,['REVIEWS']]*-1
```

And search by an specificic value of a column, for example: `Author==JJ Smith`

```python
df_books.loc[:,['AUTHOR']]=='JJ Smith'
```

## [`iloc`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.iloc.html) (Index location)

It is used to acced to a dataframe as a matrix

```python
#All results
df_books.iloc[:]
```

```python
# column INDEX search
df_books.iloc[:,0:3]
```

```python
df_books.iloc[1,3]*-1
```
