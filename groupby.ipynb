# Groupby
We group by features

image.png

```python
import pandas as pd
```

```python
path = '/content/bestsellers-with-categories.csv'
df_books = pd.read_csv(path,sep=',',header=0)
df_books.head(2)
```

```python
authors = df_books.groupby('Author')
authors
```

```python
authors.mean()
```

```python
df_books.count()
```

```python
df_books.sum()
```

```python
authors = df_books.groupby('Author').sum().loc['William Davis']
authors
```

```python
authors = df_books.groupby('Author').sum().reset_index()
authors
```

```python
df_books.groupby('Author').agg(['min','max'])
```

```python
df_books.groupby('Author').agg({'Reviews':['min','max'], 'User Rating':'sum'})
```

```python
df_books.groupby(['Author','Year']).count()
```
